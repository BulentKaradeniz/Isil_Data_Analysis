#  Data  kolonuna göre vardiya atama

```python



def vardiya_belirle(saat):
    if pd.Timestamp('00:00:00').time() <= saat < pd.Timestamp('08:00:00').time():
        return 1
    elif pd.Timestamp('08:00:00').time() <= saat < pd.Timestamp('16:00:00').time():
        return 2
    elif pd.Timestamp('16:00:00').time() <= saat <= pd.Timestamp('23:59:59').time():
        return 3

    else:
        return 0

# Vardiya sütununu ekle
df_mak_dev1['Vardiya'] = df_mak_dev1['Date'].dt.time.apply(vardiya_belirle)



# Sonucu göster
print(df_mak_dev1)
```
